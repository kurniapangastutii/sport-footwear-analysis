```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
sport_footwear = pd.read_csv('/content/drive/MyDrive/raw_data/global_sports_footwear_sales_2018_2026_ok.csv')
```

```python
#read dataset
sport_footwear.head(5)
sport_footwear.shape
sport_footwear.columns
sport_footwear.info()
sport_footwear.describe()
sport_footwear.isnull().sum()
```

```python
#changing data type
sport_footwear['order_date'] = pd.to_datetime(sport_footwear['order_date'])
sport_footwear.dtypes
```

```python
#filtering data only until 2025
sales_2018_2025 = sport_footwear[sport_footwear['order_date'].dt.year <= 2025]
```

```python
#EDA
total_revenue = sales_2018_2025['revenue_usd'].sum()
print(total_revenue)

total_units_sold = sales_2018_2025['units_sold'].sum()
print(total_units_sold)

average_price = sales_2018_2025['final_price_usd'].mean()
print(average_price)

revenue_year = sales_2018_2025.groupby(sales_2018_2025['order_date'].dt.year)['revenue_usd'].sum()
print(revenue_year)

yoy = revenue_year.pct_change() * 100
print(yoy)

revenue_vs_brand = sales_2018_2025.groupby('brand')['revenue_usd'].sum()
print(revenue_vs_brand)

categories_sold = sales_2018_2025.groupby('category')['units_sold'].sum()
print(categories_sold)

discount_sold = sales_2018_2025.groupby('discount_percent')['units_sold'].sum()
print(discount_sold)

channel_sold = sales_2018_2025.groupby('sales_channel')['units_sold'].sum()
print(channel_sold)

sales_country = sales_2018_2025.groupby('country').agg(
    total_revenue=('revenue_usd', 'sum'),
    total_units_sold=('units_sold', 'sum'),
    total_orders = ('order_id', 'nunique')
)
print(sales_country)
```
